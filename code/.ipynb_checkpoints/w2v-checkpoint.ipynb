```python
import gensim
m = "C:/Users/qwe/Documents/HSE/Coursework/code/ruscorpora_mystem_cbow_300_2_2015.bin.gz"
```

```python
if m.endswith('.vec.gz'):
    model = gensim.models.KeyedVectors.load_word2vec_format(m, binary=False)
elif m.endswith('.bin.gz'):
    model = gensim.models.KeyedVectors.load_word2vec_format(m, binary=True)
else:
    model = gensim.models.KeyedVectors.load(m)
model.init_sims(replace=True)
```

```python
def vw(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        words = f.read().split('\n')
    words = [(word + '_S') for word in words]
    cats = ['родственник_S','свойство_S', 'деятель_S', 'действие_S', 'организация_S', 'предмет_S', 'человек_S', 'агент_S',
            'учреждение_S', 'группа_S']
    res = []
    dict = {}
    for i in words[2:]:
        for cat in cats:
            try:
                dict[cat] = model.similarity(cat, i)
            except:
                dict[cat] = 0
        res.append([i, max(dict, key=dict.get)])
    print(res)
```

```python
import csv

data = []

cats = ['родственник_S','свойство_S', 'деятель_S', 'действие_S', 'организация_S', 'предмет_S', 'человек_S', 'агент_S',
            'учреждение_S', 'группа_S']

with open('C:/Users/qwe/Documents/HSE/Coursework/code/data/'
          'Data-set_normalized.csv', newline='', encoding='utf8') as f:
    reader = csv.DictReader(f, fieldnames=('№','Head', 'Gen', 'HeadNorm', 'GenNorm', 'HeadForm', 'GenForm', 'Rel',
                                               'HeadSem', 'GenSem', 'SetRel'))
    for row in reader:
        pair = {'№': row['№'], 'Head': row['Head'], 'Gen': row['Gen'].lower(),
                    'HeadNorm': row['HeadNorm'], 'GenNorm': row['GenNorm'],
                    'HeadForm': row['HeadForm'], 'GenForm': row['GenForm'],
                    'Rel': row['Rel'], 'HeadSem': '', 'GenSem': '', 'HeadSemVW': '', 'GenSemVW': '', 'SetRel': ''}
        data.append(pair)
    data = data[1:]
res = []
for p in range(len(data)):
    hdict = {}
    gdict = {}
    h = data[p]['HeadNorm']
    g = data[p]['GenNorm']
    for cat in cats:
        try:
            hdict[cat] = model.similarity(cat, h + '_S')
            #print(h, cat, model.similarity(cat, h + '_S'))
        except:
            hdict[cat] = 0
        try:
            gdict[cat] = model.similarity(cat, g + '_S')
        except:
            gdict[cat] = 0  
    try:
        hm = max(hdict, key=hdict.get)
    except:
        hm = 'other'
    try:
        gm = max(gdict, key=gdict.get)
    except:
        gm = 'other'
    res.append([h, g, hm, gm])
    data[p]['HeadSemVW'] = hm
    data[p]['GenSemVW'] = gm
print(res)

  
    #data[p]['SetRelVW'] = max(dict, key=dict.get)
with open("C:/Users/qwe/Documents/HSE/Coursework/code/data/resulttmp.csv", "w", newline='') as f:
    header = ['Вершина', 'Зависимое', 'Лемма вершины', 'Лемма зависимого', 'Форма вершины', 'Форма зависимого',
                'Отношение', 'Гиперонимы вершины', 'Гиперонимы зависимого','HSWV', 'GSWV', 'Ярлык']
    writer = csv.writer(f, delimiter=',')
    writer.writerow(header)
    for i in range(len(data)):
            writer.writerow([data[i][key] for key in data[i]])

    
    
    
```
