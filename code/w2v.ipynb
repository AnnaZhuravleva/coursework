```python
import gensim
m = "C:/Users/qwe/Documents/HSE/Coursework/code/ruscorpora_mystem_cbow_300_2_2015.bin.gz"
```

```python
if m.endswith('.vec.gz'):
    model = gensim.models.KeyedVectors.load_word2vec_format(m, binary=False)
elif m.endswith('.bin.gz'):
    model = gensim.models.KeyedVectors.load_word2vec_format(m, binary=True)
else:
    model = gensim.models.KeyedVectors.load(m)
model.init_sims(replace=True)
```

```python
with open('C:/Users/qwe/Documents/HSE/Coursework/code/sem_classes/KIN-SOC.txt', 'r', encoding='utf-8') as f:
    words = f.read().split('\n')
words = [(word + '_S') for word in words]
for i in words[2:]:
    print(model.similarity('родственник_S', i))
    print(model.similarity('свойство_S', i))
    print(model.similarity('деятель_S', i))
    print(model.similarity('действие_S', i), '\n')
```

```python
def vw(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        words = f.read().split('\n')
    words = [(word + '_S') for word in words]
    cats = ['родственник_S','свойство_S', 'деятель_S', 'действие_S',
            'организация_S', 'предмет_S', 'человек_S', 'агент_S',
            'учреждение_S', 'группа_S']
    res = []
    dict = {}
    for i in words[2:]:
        for cat in cats:
            try:
                dict[cat] = model.similarity(cat, i)
            except:
                dict[cat] = 0
        res.append([i, max(dict, key=dict.get)])
    print(res)
```
